```python
import pandas as pd
```

```python
p10_raw = pd.read_csv(r'M:\Data\GIS layers\UrbanSim smelt\2020 03 12\p10.csv',
                      usecols = ['PARCEL_ID', 'ACRES'])
print('read {} rows of parcel data'.format(p10_raw.shape[0]))
p10_raw['PARCEL_ID'] = p10_raw['PARCEL_ID'].apply(lambda x: int(round(x)))
print(p10_raw.dtypes)
```

```python
p_lookup = pd.read_csv(r'M:\Data\GIS layers\Blueprint Land Use Strategies\ID_idx\Final Blueprint\parcel_tra_hra_pda_fbp_20210816.csv')
p_lookup['pda_id_pba50_fb'].fillna('non-PDA', inplace=True)
p_lookup['pda_name'].fillna('non-PDA', inplace=True)
```

```python
p10 = p10_raw.merge(p_lookup, on='PARCEL_ID', how='left')
p10_acre = p10.groupby(['juris', 'fbp_tra_id', 'fbp_tra_cat_id', 'fbp_hra_id', 'pda_id_pba50_fb', 'pda_name'])['ACRES'].sum().reset_index()
p10_acre
```

```python
# export
p10_acre.to_csv(r'M:\Data\GIS layers\Blueprint Land Use Strategies\ID_idx\Final Blueprint\land_acreage_by_geographies.csv',
                index=False)
```
