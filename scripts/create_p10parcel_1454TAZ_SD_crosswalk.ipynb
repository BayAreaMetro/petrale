Creat a p10 / 1454TAZ / SuperDistrict crosswalk.

```python
import pandas as pd
import os
```

```python
# read TAZ-SD number crosswalk
taz_SD_num = pd.read_csv('M:/Data/GIS layers/TM1_taz/bayarea_rtaz1454_rev1_WGS84.csv',
                    usecols = ['TAZ1454', 'SUPERD'])
taz_SD_num.drop_duplicates(inplace=True)

# read SD number-name crosswalk 
sd_name = pd.read_csv(r'Z:\bayarea_urbansim\data\superdistricts.csv',
                      usecols=['number', 'name'])
sd_name.drop_duplicates(inplace=True)

# merge the two
TAZ_SD = taz_SD_num.merge(sd_name, left_on='SUPERD', right_on='number', how='left')
TAZ_SD = TAZ_SD[['TAZ1454', 'SUPERD', 'name']]
TAZ_SD.rename(columns = {'TAZ1454': 'TAZ', 'name': 'SD_NAME'}, inplace=True)

display(TAZ_SD)


# read parcel - 1454TAZ crosswalk
parcel_TAZ = pd.read_csv(r'Z:\bayarea_urbansim\data\2018_10_17_parcel_to_taz1454sub.csv',
                         usecols = ['PARCEL_ID', 'ZONE_ID'])
parcel_TAZ['PARCEL_ID'] = parcel_TAZ['PARCEL_ID'].apply(lambda x: int(round(x)))
parcel_TAZ.rename(columns={'ZONE_ID': 'TAZ',
                           'PARCEL_ID': 'parcel_id'}, inplace=True)

parcel_TAZ_SD = parcel_TAZ.merge(TAZ_SD, on='TAZ', how='left')

print(parcel_TAZ_SD.shape)
display(parcel_TAZ_SD.head())
```

```python
parcel_TAZ_SD.to_csv(r'M:\Data\Urban\BAUS\PBA50\p10_parcel_TAZ_SD.csv', index=False)
```
