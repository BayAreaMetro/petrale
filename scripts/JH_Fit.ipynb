Calculate Low-wage shares by County and MTC industry type

```python
#Load employment-based and residence-based jobs data
jobs_EB=pd.read_csv(r"C:\Users\jhalpern\Documents\R\PBA2050\Outputs\Jobs-HousingFit\PUMS2014-2018 Income by Industry and Emp County.csv")
jobs_RB=pd.read_csv(r"C:\Users\jhalpern\Documents\R\PBA2050\Outputs\Jobs-HousingFit\PUMS2014-2018 Income by Industry and Res County.csv")
```

```python
#Note uncoded are not included
jobs_EB['tot_pwgt'] = jobs_EB['1_less than 40k']+jobs_EB['2_40k to 49,999']+jobs_EB['3_50k to 79,999']+jobs_EB['4_80k to 99,999']+jobs_EB['5_100k to 149,999']+jobs_EB['6_150k to 249,999']+jobs_EB['7_250k+']
```

```python
jobs_EB['shareLW_Under40k']=jobs_EB['1_less than 40k']/jobs_EB['tot_pwgt']
jobs_EB['shareLW_Under50k']=(jobs_EB['1_less than 40k']+jobs_EB['2_40k to 49,999'])/jobs_EB['tot_pwgt']
jobs_EB['shareLW_Under80k']=(jobs_EB['1_less than 40k']+jobs_EB['2_40k to 49,999']+jobs_EB['3_50k to 79,999'])/jobs_EB['tot_pwgt']
```

```python
#There is a significant difference between using less than $40k and less than $50k
#Measure and describe the difference
jobs_EB['diff40kto50k']=jobs_EB['2_40k to 49,999']/jobs_EB['tot_pwgt']
jobs_EB[["shareLW_Under40k","shareLW_Under50k","diff40kto50k"]].describe()
```

```python
#Number of NaN (Industry) is 22,633
jobs_EB[jobs_EB['Industry'].isnull()]['tot_pwgt'].sum()
```

Group by County

```python
#Total person weights similar to Vital Signs employment numbers
jobs_county['tot_pwgt'].sum()
```

```python
#Recalculate shares using county numbers (for later comparison)
jobs_county['shareLW_Under40k']=jobs_county['1_less than 40k']/jobs_county['tot_pwgt']
jobs_county['shareLW_Under50k']=(jobs_county['1_less than 40k']+jobs_county['2_40k to 49,999'])/jobs_county['tot_pwgt']
jobs_county['shareLW_Under80k']=(jobs_county['1_less than 40k']+jobs_county['2_40k to 49,999']+jobs_county['3_50k to 79,999'])/jobs_county['tot_pwgt']
jobs_county['diff40kto50k']=jobs_county['2_40k to 49,999']/jobs_county['tot_pwgt']
jobs_county[["shareLW_Under40k","shareLW_Under50k","diff40kto50k"]].describe()
```

## Combine shares and juris/county summary

```python
#Before merging, need to pivot Industry from rows to columns
jobs_EB.rename(inplace=True,columns={"shareLW_Under40k": "Under40k", "shareLW_Under50k": "Under50k","shareLW_Under80k": "Under80k"})
jobs_pivot=jobs_EB.pivot(index='County_Emp', columns='Industry', values=['Under40k','Under50k','Under80k'])
```

```python
jobs_pivot
```

```python
jobs_pivot=jobs_pivot.reset_index(drop=False)
```

```python
len(jobs_pivot.columns)
```

```python
tempcolumn=['County_Emp']+[x+y for x in ['U40k_','U50k_','U80k_'] for y in ['NaN','AGREMPN','FPSEMPN','HEREMPN','MWTEMPN','OTHEMPN','RETEMPN']]
```

```python
print(tempcolumn)
```

```python
len(tempcolumn)
```

```python
jobs_pivot.columns=tempcolumn
```

```python
jobs_pivot
```

```python
juris15_shares = pd.merge(juris2015, jobs_pivot,  how='left', left_on=['County'], right_on = ['County_Emp'])
```

```python
#number of low wage jobs for each HH income cutoff
juris15_shares['LWJobs_U40k']=round(juris15_shares['U40k_AGREMPN']*juris15_shares['agrempn']+ \
                            juris15_shares['U40k_MWTEMPN']*juris15_shares['mwtempn']+ \
                            juris15_shares['U40k_RETEMPN']*juris15_shares['retempn']+ \
                            juris15_shares['U40k_FPSEMPN']*juris15_shares['fpsempn']+ \
                            juris15_shares['U40k_HEREMPN']*juris15_shares['herempn']+ \
                            juris15_shares['U40k_OTHEMPN']*juris15_shares['othempn'])

juris15_shares['LWJobs_U50k']=round(juris15_shares['U50k_AGREMPN']*juris15_shares['agrempn']+ \
                            juris15_shares['U50k_MWTEMPN']*juris15_shares['mwtempn']+ \
                            juris15_shares['U50k_RETEMPN']*juris15_shares['retempn']+ \
                            juris15_shares['U50k_FPSEMPN']*juris15_shares['fpsempn']+ \
                            juris15_shares['U50k_HEREMPN']*juris15_shares['herempn']+ \
                            juris15_shares['U50k_OTHEMPN']*juris15_shares['othempn'])

juris15_shares['LWJobs_U80k']=round(juris15_shares['U80k_AGREMPN']*juris15_shares['agrempn']+ \
                            juris15_shares['U80k_MWTEMPN']*juris15_shares['mwtempn']+ \
                            juris15_shares['U80k_RETEMPN']*juris15_shares['retempn']+ \
                            juris15_shares['U80k_FPSEMPN']*juris15_shares['fpsempn']+ \
                            juris15_shares['U80k_HEREMPN']*juris15_shares['herempn']+ \
                            juris15_shares['U80k_OTHEMPN']*juris15_shares['othempn'])
```

```python
#Test Bay area J-HH fit 
#It should not be this low... .73
juris15_shares['LWJobs_U50k'].sum()/juris15_shares['hhincq1'].sum()
```

```python
#Test Bay area J-HH fit 
#Slightly higher if using Q1+Q2 and Under $80k annual HH income
juris15_shares['LWJobs_U80k'].sum()/(juris15_shares['hhincq1'].sum()+juris15_shares['hhincq2'].sum())
```

```python
# ~13% of all jobs are low wage
#this number also seems low
juris15_shares['LWJobs_U50k'].sum()/juris15_shares['totemp'].sum()
```

```python
juris15_shares['Jobs-HH Fit']=juris15_shares['LWJobs_U50k']/juris15_shares['hhincq1']
```
