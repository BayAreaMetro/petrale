```python
import os
import pandas as pd
```

```python
folder = 'C:\\Users\\{}\\Box\\Modeling and Surveys\\Urban Modeling\\Bay Area UrbanSim\\PBA50'.format(os.getenv('USERNAME'))
```

```python
np_15 = pd.read_csv(os.path.join(folder, 'EIR runs', 'Baseline Large (s25) runs', 'NP_v8_FINAL',
                               'run314_county_summaries_2015.csv'),
                    usecols = ['COUNTY_NAME', 'TOTEMP', 'TOTHH'])
np_15['ratio_2015'] = np_15['TOTEMP'] / np_15['TOTHH']
display(np_15.sort_values(by='COUNTY_NAME'))
np_15.set_index('COUNTY_NAME', inplace=True)
```

```python
np_50 = pd.read_csv(os.path.join(folder, 'EIR runs', 'Baseline Large (s25) runs', 'NP_v8_FINAL',
                               'run314_county_summaries_2050.csv'),
                    usecols = ['COUNTY_NAME', 'TOTEMP', 'TOTHH'])
np_50['ratio_np'] = np_50['TOTEMP'] / np_50['TOTHH']
display(np_50.sort_values(by='COUNTY_NAME'))
np_50.set_index('COUNTY_NAME', inplace=True)
```

```python
fbp = pd.read_csv(os.path.join(folder, 'Final Blueprint runs', 'Final Blueprint (s24)','BAUS v2.25 - FINAL VERSION',
                               'run182_county_summaries_2050.csv'),
                  usecols = ['COUNTY_NAME', 'TOTEMP', 'TOTHH'])
fbp['ratio_fbp'] = fbp['TOTEMP'] / fbp['TOTHH']
display(fbp.sort_values(by='COUNTY_NAME'))
fbp.set_index('COUNTY_NAME', inplace=True)
```

```python
alt1 = pd.read_csv(os.path.join(folder, 'EIR runs', 'Alt1 (s26) runs', 'Alt1_v3_test_far_tiers_FINAL_EIR_ALT',
                               'run375_county_summaries_2050.csv'),
                   usecols = ['COUNTY_NAME', 'TOTEMP', 'TOTHH'])
alt1['ratio_alt1'] = alt1['TOTEMP'] / alt1['TOTHH']
display(alt1.sort_values(by='COUNTY_NAME'))
alt1.set_index('COUNTY_NAME', inplace=True)
```

```python
alt2 = pd.read_csv(os.path.join(folder, 'EIR runs', 'Alt2 (s28) runs', 'Alt2_v1_FINAL_EIR_ALT',
                               'run374_county_summaries_2050.csv'),
                   usecols = ['COUNTY_NAME', 'TOTEMP', 'TOTHH'])
alt2['ratio_alt2'] = alt2['TOTEMP'] / alt2['TOTHH']
display(alt2.sort_values(by='COUNTY_NAME'))
alt2.set_index('COUNTY_NAME', inplace=True)
```

```python
summary = pd.concat([np_15[['ratio_2015']],
                     np_50[['ratio_np']],
                     fbp[['ratio_fbp']],
                     alt1[['ratio_alt1']],
                     alt2[['ratio_alt2']]
                    ], axis=1, join="inner")
summary.sort_values(by='COUNTY_NAME')
```
