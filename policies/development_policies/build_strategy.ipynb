### inclusionary housing setting

```python
import pandas as pd
import numpy as np
df = pd.read_csv(r'C:\Users\ywang\Box\Modeling and Surveys\Urban Modeling\Bay Area UrbanSim\PBA50\Current PBA50 Large General Input Data\2020_11_10_parcels_geography.csv')

# build lists of fbpchcat with high- and medium-level inclusionary requirements
high = df.loc[(df.fbp_gg_id == 'GG') & df.fbp_tra_id.notnull() & (
        (df.fbp_sesit_id == 'HRADIS') | (df.fbp_sesit_id == 'HRA'))]['fbpchcat'].unique()

med = df.loc[((df.fbp_gg_id == 'GG') & (df.fbp_tra_id == 'tra1') & ((df.fbp_sesit_id == 'DIS') | df.fbp_sesit_id.isnull())
        ) | (
        (df.fbp_gg_id == 'GG') & ((df.fbp_tra_id == 'tra2a') | (df.fbp_tra_id == 'tra2b') | (df.fbp_tra_id == 'tra2c')) & ((df.fbp_sesit_id == 'DIS') | df.fbp_sesit_id.isnull())
        ) | (
        (df.fbp_gg_id == 'GG') & df.fbp_tra_id.isnull() & ((df.fbp_sesit_id == 'HRADIS') | (df.fbp_sesit_id == 'HRA')))]['fbpchcat'].unique()

display('{} fbpchcats have high-level inclusionary requirement (20%)'.format(len(high)))
display('{} fbpchcats have medium-level inclusionary requirement (15%)'.format(len(med)))
display('{} fbpchcats have low-level inclusionary requirement (10%)'.format(len(df.fbpchcat.unique()) - len(high) - len(med)))

# merge the two lists with the entire fbpchcat
fbpchcat_df = df[['fbpchcat']].drop_duplicates()
#display(fbpchcat_df.head())

high_data = {'fbpchcat':list(high), 'level_h':['high'] * len(high)}
high_df = pd.DataFrame(high_data)

medium_data = {'fbpchcat':list(med), 'level_m':['medium'] * len(med)}
medium_df = pd.DataFrame(medium_data)

# label fbpchcat with low-level inclusionary requirements
fbpchcat_df = fbpchcat_df.merge(high_df, on='fbpchcat', how = 'left').merge(medium_df, on='fbpchcat', how='left')
fbpchcat_df['level_l'] = ''
fbpchcat_df.loc[fbpchcat_df.level_h.isnull() & fbpchcat_df.level_m.isnull(), 'level_l'] = 'low'
fbpchcat_df['fbpchcat'] = fbpchcat_df['fbpchcat'].apply(lambda x: '      - '+x)
display(fbpchcat_df)

# export
fbpchcat_df.to_csv('inclusionary_fbpchcat.csv')
```
